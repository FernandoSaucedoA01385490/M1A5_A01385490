```python
# 1.
ventas = pandas.read_csv('sales_data_types.csv')
ventas.head()
```

```python
# 2.
ventas = ventas.rename(columns={
    "Customer Number": "Customer_Number",
    "Customer Name": "Customer_Name",
    "Percent Growth": "Percent_Growth",
    "Jan Units": "Jan_Units",
    "Shipping Cost": "Shipping_Cost",
    "Order Priority": "Order_Priority"
})
ventas.head()
```

```python
# 3.
print("Order_ID debería ser un tipo de dato numérico o categórico, ")
print("Order_Date, Ship_Date deberían ser fechas.")
print("Sales, Quantity, Discount, Profit, Shipping Cost deberían ser numéricos.")
print("")
print(ventas.dtypes)
```

```python
# 4.
print("No se pueden sumar las ventas porque las columnas son de tipo 'object'")
print("Se deben convertir las columnas 2016 y 2017 a tipo numérico (float64)")
print("")

ventas["2016"] = ventas["2016"].replace({r'[$,]': ''}, regex=True).astype("float64")
ventas["2017"] = ventas["2017"].replace({r'[$,]': ''}, regex=True).astype("float64")

ventatotal = ventas['2016'].sum() + ventas['2017'].sum()

print(f"Las ventas totales de 2016 a 2017 son: {ventatotal}")
```

```python
# 5.
ventas["Customer_Number"] = ventas["Customer_Number"].astype("int64")
ventas["Customer_Name"] = ventas["Customer_Name"].astype("string")
ventas["Percent_Growth"] = ventas["Percent_Growth"].str.replace('%', '').astype("float64")
ventas["Jan_Units"] = pandas.to_numeric(ventas["Jan_Units"], errors='coerce')
ventas["Active"] = ventas["Active"].map({"Y": True, "N": False})

print(ventas.dtypes)
ventas.head()
```

```python
# 6.
ventas['Start_date'] = pandas.to_datetime(ventas[['Month', 'Day', 'Year']])
ventas.head()
```

```python
# 7.
print("7.1")
print(ventas.iloc[:3])
print("")
print("7.2")
print(ventas.iloc[:2, :2])
```

```python
# 8.
ventas.sort_values(by='2016')
```

3. Genera un boxplot para cada columna numérica. Usa la función de "seaborn.boxplot"
4. Genera un boxplot de "sepal_width" agrupada por especies. Para cada boxplot calcula los valores máximos, mínimos, Q1, Q3 y mediana
5. En la boxplot anterior, sobrepon un swarmplot para mirar individualmente los puntos de datos.
6. Genera un panel con 4 figuras usando la función de subtablas "matplotlib.pyplot". Cada panel debe mostrar un boxplot por especies para cada una de las 4 variables numéricas (longitud y ancho del sépalo, y longitud y ancho del pétalo).
7. ¿Qué variable separa mejor a las especies?
8. ¿Qué variable(s) muestra(n) valores atípicos? ¿En qué especies?

```python
# 1.
plantas = pandas.read_csv('iris.csv')
plantas.head()
```

```python
# 2.
plantas["variety"] = plantas["variety"].astype("string")

plantas.dtypes
```

```python
# 3.
for column in plantas.select_dtypes(include=['number']):
    plt.figure()
    sns.boxplot(x=plantas[column])
    plt.title(f'Boxplot de {column}')
    plt.xlabel(column)
    plt.show()
```

```python
# 4.
for column in plantas.select_dtypes(include=['number']):
    plt.figure()
    sns.boxplot(x=plantas["variety"], y=plantas[column])
```

```python
# 5.
for column in plantas.select_dtypes(include=['number']):
    plt.figure()
    sns.boxplot(x=plantas["variety"], y=plantas[column], color="skyblue", width=0.5)
    sns.swarmplot(x=plantas["variety"], y=plantas[column], color="black", alpha=0.7)
    plt.title(f'Boxplot y Swarmplot de {column}', fontsize=14, fontweight='bold')
    plt.xlabel("Variety", fontsize=12)
    plt.ylabel(column, fontsize=12)
    plt.show()
```

```python
# 6.
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for idx, column in enumerate(plantas.select_dtypes(include=['number'])):
    sns.boxplot(x=plantas["variety"], y=plantas[column], color="skyblue", width=0.5, ax=axes[idx])
    sns.swarmplot(x=plantas["variety"], y=plantas[column], color="black", alpha=0.7, ax=axes[idx])
    axes[idx].set_title(f'Boxplot y Swarmplot de {column}', fontsize=14, fontweight='bold')
    axes[idx].set_xlabel("Variety", fontsize=12)
    axes[idx].set_ylabel(column, fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# 7.
print("La variable que mejor separa las especies es 'petal_width'")
print("Esta variable tiene la menor superposición entre las diferentes especies, y los valores atípicos son mínimos.")
```
